```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
Univ = pd.read_csv("Universities.csv")
Univ
```

# Normalization..(here i have mentioned 2 methods to do the normalization)

```python
#method 1
#we are doing normalization,it converts into[0 to 1]
def norm_func(i):
  x=(i-i.min())/(i.max()-i.min())#i means column..
  return(x)
```

```python
# Normalized data frame (considering the numerical part of data)
df_norm = norm_func(Univ.iloc[:,1:]) # from 1st column because 0 index col is univ names and we don't need that column
df_norm
```

```python
#method 2
from sklearn.preprocessing import MinMaxScaler
trans = MinMaxScaler()
data = pd.DataFrame(trans.fit_transform(Univ.iloc[:,1:]))
data
```

see both methods has same answers..

# Dendrogram

```python
from scipy.cluster.hierarchy import linkage
import scipy.cluster.hierarchy as sch #for creating dendogram
#p = np.array(df_norm) #converting into numpy array format
z=linkage(df_norm,method="average",metric="euclidean")
plt.figure(figsize=(15,8))
plt.title("Hiearchial Clustering Dendrogram")
plt.xlabel("Index")
plt.ylabel("Distance")
sch.dendrogram(z)
plt.show() #leaf_rotation =6 #rotates x axis label
#leaf_font_size=15 #font size for the x axix labels
plt.show()
```

```python
#we have 5 clusters..
```

# Model

```python
from sklearn.cluster import AgglomerativeClustering
import warnings
warnings.filterwarnings("ignore")
h_complete=AgglomerativeClustering(n_clusters=5,linkage='average',metric="euclidean").fit(df_norm) #this is our model..
cluster_labels=pd.Series(h_complete.labels_)
cluster_labels
Univ['clust']=cluster_labels #creating a new column & assigning it to new column
Univ
```

```python
Univ.iloc[:,1:].groupby(Univ.clust).mean() #grouping all ur universities
```

```python
data=Univ[(Univ.clust==1)]
data
```
