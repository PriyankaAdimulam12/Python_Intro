# Duplicates

So fare we have seen datsets like salaries, universities..
You can also create datset, so u can create using using ur "pandas library" with Dataframe..

```python
df2=pd.DataFrame({'col1':['A','A','B','B','B'],'col2':[2,2,2,3,3]})
#we created a dataframe,now we will see what is the purpose..
df2
```

```python
df2.duplicated() #this tells us the duplicated values,i.e the repeated values in or dataframe..
#false means its not duplicated,  true means which is duplicated,i.e the repeated ones
#we can say that our data has repeting values
```

```python
df2[df2.duplicated()] # this shows us the duplicated records
```

```python
#we can drop the duplicated records..
df3=df2.drop_duplicates()
df3
#output shows the record without the duplicated values
```

```python
 df3
 #this is our output right,but here the index are not in a order 0,2,3..
 #so we need to rearrange the index in a order..
```

```python
df4=df3.reset_index()
#so here we have reset the index,now you need to drop the previous index column..
```

```python
df4
```

```python
df5=df4.drop('index',axis=1)
df5
```
